# Data analysis:-

```python
print("Country with most attacks: ",terror_data["country"].value_counts().idxmax())
print("City with most attacks: ",terror_data["city"].value_counts().index[1])
print("Region with most attacks: ",terror_data["region"].value_counts().idxmax())
print("Year with most attacks: ",terror_data["year"].value_counts().idxmax())
print("Month with most attacks: ",terror_data["month"].value_counts().idxmax())
print("Group with most attacks: ",terror_data["group"].value_counts().index[1])
print("Most Attack Types: ",terror_data["attacktype"].value_counts().idxmax())
```

# Data visualization:-

### 1.Top Countries Affected due to terrorism

```python
plt.figure(figsize=(15,6))
sns.barplot(terror_data['country'].value_counts()[:10].index,terror_data['country'].value_counts()[:10])
plt.title('Top Countries Affected',fontsize=20)
plt.xlabel('Countries',fontsize = 15)
plt.ylabel('Count',fontsize = 15)
plt.xticks(rotation = 45)
plt.show()
```

### 2. Total number of attacks per years

```python
plt.figure(figsize = (15,6))
year = terror_data['year'].unique()
years_count = terror_data['year'].value_counts(dropna = False).sort_index()
sns.barplot(x = year, y = years_count)
plt.title('Attacks In Years',fontsize = 20)
plt.xlabel('Attacking Year',fontsize = 15)
plt.ylabel('Number of Attacks Each Year',fontsize = 15)
plt.xticks(rotation = 90)

plt.show()
```

### 3. Top cities affected due to terrorism

```python
plt.figure(figsize=(15,6))
sns.barplot(terror_data['city'].value_counts()[:10].index,terror_data['city'].value_counts()[:10].values)
plt.title('Top Cities Affected',fontsize = 20)
plt.xlabel('City',fontsize = 15)
plt.ylabel('Count',fontsize = 15)
plt.xticks(rotation = 45)
plt.show()
```

### 4.Top Most number of people killed per year

```python
plt.figure(figsize = (15,6))
terror_data.groupby(['year'])['killed'].sum().sort_values(ascending = False).head(20).plot(kind = 'bar')
plt.xticks(rotation=45)
plt.title('No. of people killed Year wise',fontsize = 20)
plt.ylabel("Killed",fontsize = 15)
plt.xlabel("Year",fontsize = 15)
plt.show()
```

### 5.Attacktype used for terrorism

```python
plt.figure(figsize = (12,6))

sns.barplot(x = terror_data['attacktype'].value_counts()[1:11].values, y = terror_data['attacktype'].value_counts()[1:11].index)
plt.xticks(rotation = 0)
plt.xlabel("Attacktype",fontsize=15)
plt.ylabel("Name of attack",fontsize=15)
plt.title("No. of attacktype",fontsize=20)
plt.show()
```

### 6. Top countries where no. of people killed

```python
plt.figure(figsize = (15,6))
x = terror_data.groupby(['country'])['killed'].sum().sort_values(ascending = False).head(10).plot(kind = 'bar')
plt.xticks(rotation=0)
plt.title('No. of people killed Country wise',fontsize = 20)
plt.ylabel("NO. of persons Killed",fontsize = 15)
plt.xlabel("Countries",fontsize = 15)
plt.show()
```

### 7.Top region where no. of people killed

```python
plt.figure(figsize = (15,6))
terror_data.groupby(['region'])['killed'].sum().sort_values(ascending = False).head(10).plot(kind = 'bar', colormap = 'PRGn')
plt.xticks(rotation=45)
plt.title('No. of people killed Region wise',fontsize = 20)
plt.ylabel("No. of people Killed",fontsize = 15)
plt.xlabel("Regions",fontsize = 15)
plt.show()
```

### 8. Terrorist Activities By Region per Year

```python
pd.crosstab(terror_data['year'], terror_data['region']).plot(kind = 'area', stacked = False, figsize = (15,6))
plt.title('Terrorist Activities By Region In Each Year', fontsize = 20)
plt.ylabel('Number of Attacks', fontsize = 15)
plt.xlabel("Year", fontsize = 15)
plt.show()
```

### 9.Top Terrorist Organisation with Highest Terror Attacks

```python
plt.figure(figsize = (12,6))
sns.barplot(x = terror_data['group'].value_counts()[1:11].values, y = terror_data['group'].value_counts()[1:11].index)
plt.title('Top 10 Terrorist Organisation with Highest Terror Attacks',fontsize = 20)
plt.ylabel('Terror Group Name',fontsize = 15)
plt.xlabel('Number of Attacks',fontsize = 15)
plt.xticks(rotation= 0)
plt.show()
```
